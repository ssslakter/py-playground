```
import numpy as np
import matplotlib.pyplot as plt
```

```
d = 2
x = np.random.randn(d, int(1e6))
```

```
data, x_e, y_e = np.histogram2d(*x, bins=50)
plt.imshow(np.histogram2d(*x, bins=50)[0], extent=[x_e[0], x_e[-1], y_e[0], y_e[-1]])
```

```
l = [100]
l.extend(range(1, 20, 2))
for d in l:
    x = np.random.randn(d, int(1e6))
    norms = np.linalg.norm(x, axis=0)
    data, x_e = np.histogram(norms,bins=70, density=True)
    plt.plot(x_e[:-1], data, label=d, color = plt.colormaps['magma'](d/21))
plt.legend()
```
