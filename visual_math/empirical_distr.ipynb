# Empirical Distribution Visualization
This script demonstrates the process of generating random data, computing cumulative sums for integration, and plotting an empirical distribution against a theoretical distribution using Python's `numpy` and `matplotlib` libraries.


```python
import numpy as np
import matplotlib.pyplot as plt
from functools import partial
```

```python
X = np.random.randn(100)
```

```python
def integrate(y,x):
    deltas = x[1:]-x[:-1]
    return np.cumsum(y)[1:]*deltas
```

```python
x= np.linspace(-5,5,100)
plt.plot(x[1:],integrate(x,x))
```
