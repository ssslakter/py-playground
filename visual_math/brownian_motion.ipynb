```python
import matplotlib.pyplot as plt
import numpy as np
import torch
import torch.nn.functional as F
```

Normally distributed variables $x(t)\sim \mathcal{N}(0, 1)$ for all $t$

```python
n = 100
T = 10
t = torch.linspace(0, T, n*T)
x = torch.randn(n*T)
plt.plot(t,x)
```

Weiner process $W_t$ describing Brownian motion is characterized by four facts:
- $W_0 = 0$
- $W_t$ is almost surely continuous
- $W_t$ has independent increments
- ${\displaystyle W_{t}-W_{s}\sim {\mathcal {N}}(0,t-s)}$ (for  ${\displaystyle 0\leq s\leq t}$).

```python
dt = 1/n
w = torch.cumsum(torch.randn(n*T)*np.sqrt(dt), 0)
w[0]=0 # by definition of Wiener process
plt.plot(t,w)
```

```python
ax = plt.subplots(3, 3, figsize=(10,10))[1]
scales = [1, 5, 10, 30, 50, 100, 500, 1000, 5000]
for i, scale in enumerate(scales):
    ax[i//3, i%3].plot(torch.linspace(0, T, scale*T), torch.cumsum(torch.randn(scale*T)*np.sqrt(1/scale), 0))
```

From definition it follows that $W_t \sim \mathcal{N}(0, t)$.

But it doesn't work since if $x_t \sim \mathcal{N}(0, 1)$ then $Var[x_t - x_s]=t^2+s^2$ and not $t-s$.
